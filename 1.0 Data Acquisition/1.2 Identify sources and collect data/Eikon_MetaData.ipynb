```python
securities = {}

class Security:
    def __init__(self, name, ticker, isin=None):
        self.name = name
        self.ticker = ticker
        self.isin = isin
    @property
    def symbol(self):
        return self.ticker.replace("TH:","")
    @property
    def ric(self):
        return self.ticker

for i in range(2, sheet.nrows):
    r = sheet.row(i)
    print("{} - {}".format(r[3].value,r[1].value))
    ticker = r[3].value
    name = r[1].value
    isin = r[5].value
    securities[ticker] = Security(name=name, ticker=ticker, isin=isin)
```

```python
len(securities)
```

```python
sheet.row(603)
```

```python
import pandas as pd
df = pd.DataFrame({})
for k, v in securities.items():
    r = {'symbol': v.symbol, 'name': v.name, 'RIC': v.ric, 'ISIN': v.isin}
    df = df.append(r, ignore_index=True)
    print(v.symbol, v.name)
```
