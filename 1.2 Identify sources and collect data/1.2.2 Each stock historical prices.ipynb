```python
import pandas as pd
import pandas_datareader as pdr
import datetime as dt
import os, os.path
```

```python
## Load ticker list
```

```python
ticker_list = "stock_ticker.csv"
stocks = pd.read_csv(ticker_list)
stocks
```

## Set study period

```python
start = dt.datetime(2014,1,1)
```

## Download stock data from Yahoo

```python
if not os.path.exists("YahooFinance"):
    os.mkdir('YahooFinance')
    print("Data directory created.")
```

```python
for s in stocks['symbol'].values:
    print(s)
    try:
        data = pdr.get_data_yahoo('%s.BK' % s,start=start)
        outfile = "YahooFinance/%s.csv" % s
        data.to_csv(outfile)
    except Exception as e:
        print("Error: %s %s" % (s, str(e)))
```
