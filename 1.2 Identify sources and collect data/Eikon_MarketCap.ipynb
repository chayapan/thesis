# Time-Series of Market cap of the Universe

```python

series = {}

# Date column is column 0
# Convert spreadsheet date value to python date
s1 = pd.Series(sheet.col(0)[4:]) # observation starts at row 4
s1 = s1.apply(lambda x: datetime(*xlrd.xldate_as_tuple(x.value, wb.datemode)))

for i in range(1, sheet.ncols):
    c = sheet.col(i)
    print("{} -".format(c[3].value))
    name = c[3].value
    # df = pd.DataFrame()
    if not "#ERROR" in name:
        try:
            # Convert spreadsheet market cap value to python
            s2 = pd.Series(c[4:]) # observation starts at row 4
            s2 = s2.apply(lambda x: x.value)

            df = pd.concat([s1, s2], axis=1)
            df.columns = ['Date', 'Market Value']
            df = df.set_index('Date')

            # Add to market dictionary dataset
            series[name] = df
        except Exception as e:
            print("Error for %s %s" % (name, str(e)))
```

```python
len(series)
```

```python
for k in series:
    df = series[k]
    df.plot(title="Market cap %s" % k)
```
