# Return - Holding period return


Holding Period Return (HPR) and annualized HPR for returns over multiple years

https://www.investopedia.com/terms/h/holdingperiodreturn-yield.asp

Holding Period Return

= ( Income + (End of Period Value - Initial Value) ) / Initial Value


Holding period return is thus the total return received from holding an asset or portfolio of assets over a specified period of time, generally expressed as a percentage.


1. Holding period return (or yield) is the total return earned on an investment during the time that it has been held.
2. A holding period is the amount of time the investment is held by an investor, or the period between the purchase and sale of a security.
3. Holding period return is useful for making like comparisons between returns on investments purchased at different periods in time.

```python
t = 10 # holding period in days
y = 15 # yield in percentage
p = 100 # principal amount

def holding_period_return_est(principal=p, pct_yield=y, num_periods=t):
    """Holding period return function 
       estimates the amount to receive after the end of holding period."""
    hpr = 0
    income = principal * pct_yield / 100.
    days = num_periods
    rate_of_return = pct_yield / 100. #
    start_amt = principal
    end_amt = start_amt * (1. + rate_of_return)
    hpr = income + (end_amt - start_amt) / start_amt
    return hpr

holding_period_return_est(p, y, t)
```
